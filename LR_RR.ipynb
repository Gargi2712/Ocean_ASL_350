```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("temp.csv")
# Convert Date to datetime format (if not already)
df['Date'] = pd.to_datetime(df['Date'])

# Extract features from the Date (you can use Year, Month, etc.)
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Feature and target variable
X = df[['Year', 'Month', 'Day']]  # Independent variables (date features)
y = df['Temp']                    # Dependent variable (temperature)
```

# For Temp, Using Linear Regression

```python
# Train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X, y)

# Predict for the next 5 years (we'll create future dates)
future_dates = pd.date_range(start=df['Date'].max(), periods=60, freq='M')  # 5 years ahead
future_df = pd.DataFrame(future_dates, columns=['Date'])
future_df['Year'] = future_df['Date'].dt.year
future_df['Month'] = future_df['Date'].dt.month
future_df['Day'] = future_df['Date'].dt.day

# Predict temperature using the trained model
forecast_temp_lr = lr_model.predict(future_df[['Year', 'Month', 'Day']])

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Temp'], label='Historical Data')
plt.plot(future_dates, forecast_temp_lr, label='Predicted Data (Linear Regression)', color='red')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Forecast (Linear Regression)')
plt.legend()
plt.show()

# Print predicted temperatures for the next 5 years
print(forecast_temp_lr)
```

Year captures long-term trends e.g., global warming
Month can represent seasonal variations
Temperature can also follow daily patterns, especially in cases of daily temperature fluctuations

Temperature data often follows cyclical patterns, with months reflecting seasonal variations (e.g., higher temperatures in summer months like June-August compared to winter months like January-December). Daily fluctuations can also impact temperature, while the year variable helps capture long-term trends, such as global warming. By using separate features like month, day, and year, the model can effectively capture both cyclical and long-term patterns in temperature.

# For Temp, Using Random Forest

```python
# Train the RandomForest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Generate future dates for prediction (next 5 years)
future_dates = pd.date_range(df['Date'].max(), periods=60, freq='M')  # 60 months = 5 years
future_dates_df = pd.DataFrame(future_dates, columns=['Date'])

# Extract cyclical features for future dates
future_dates_df['Year'] = future_dates_df['Date'].dt.year
future_dates_df['Month'] = future_dates_df['Date'].dt.month
future_dates_df['Day'] = future_dates_df['Date'].dt.dayofyear

# Predict temperatures for the next 5 years
forecast_temp_rf = rf_model.predict(future_dates_df[['Year', 'Month', 'Day']])

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Temp'], label='Historical Data')
plt.plot(future_dates, forecast_temp_rf, label='Predicted Data (Random Forest)', color='green')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Forecast (Random Forest)')
plt.legend()
plt.show()

# Print predicted temperatures for the next 5 years
print(forecast_temp_rf)
```
